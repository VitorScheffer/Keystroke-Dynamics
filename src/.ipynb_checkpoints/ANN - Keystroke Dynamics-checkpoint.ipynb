# ANN

```python
import numpy as np
import pandas as pd
from keystroke_dynamics import get_dados
from keystroke_dynamics import learn_curve


from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix

from sklearn.metrics import classification_report
```

```python
# busca dados
X, y, X_train, X_test, y_train, y_test, y_labels = get_dados()
```

```python
classificador = MLPClassifier(random_state=42,hidden_layer_sizes=(84, ),max_iter=600,activation= 'relu', learning_rate= 'invscaling', solver='adam')
```

```python
classificador.fit(X_train,y_train)
```

```python
# predict the output using the test data on the learned model
y_predict = classificador.predict(X_test)
model_accuracy = metrics.accuracy_score(y_test, y_predict)
print('Acurácia do modelo:',model_accuracy)
```

```python
print(classification_report(y_test, y_predict, target_names=y_labels))
```

```python
learn_curve(classificador, "ANN", X, y).show()
```
